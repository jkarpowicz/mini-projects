### Data Preprocessing

```python
profiles.columns
```

```python
profiles.isnull().sum()
```

```python
cols = ['diet', 'orientation', 'drinks', 'ReligionHL',
       'sex', 'job', 'education']
df = profiles[cols].dropna()
df.shape
```

```python
for col in cols[1:]:
    df = pd.get_dummies(df, columns=[col], prefix = [col])
```

```python
df.head()
```

### Machine Learning

Features prediction body type

```python
labels = df[['diet']]
features_data = df[df.columns[1:]]
#x = features_data.values
#min_max_scaler = MinMaxScaler()
#x_scaled = min_max_scaler.fit_transform(x)
```

```python
labels.shape
```

```python
features_data.shape
```

```python
## SPlitting data into train/test
X_train, X_test, y_train, y_test = train_test_split(features_data, labels, train_size=0.8, test_size=0.2, random_state = 100)
```

```python
# Predicting body type from smoke and drugs abuse using Naive Bayes Classifier
classifier = MultinomialNB()
classifier.fit(X_train, y_train)
NB_predictions = classifier.predict(X_train)
print(classifier.score(X_test, y_test))
```

```python
print(classification_report(y_train, NB_predictions))
```

```python
classifier1 = KNeighborsClassifier(n_neighbors = 10)
classifier1.fit(X_train, y_train)
KNN_predictions = classifier1.predict(X_train)
print(classifier1.score(X_test, y_test))
print(classification_report(y_train, KNN_predictions))
```

```python
classifier2 = DecisionTreeClassifier()

classifier2.fit(X_train, y_train) 
DT_predictions = classifier2.predict(X_train)
print(classifier2.score(X_test, y_test))
print(classification_report(y_train, DT_predictions))
```

```python
classifier3 = LogisticRegression()
classifier3.fit(X_train, y_train)
LR_predictions = classifier3.predict(X_train)
print(classifier.score(X_test, y_test))
print(classification_report(y_train, LR_predictions))
```
